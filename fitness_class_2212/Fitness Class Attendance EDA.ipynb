Task 2

創建一個可視化效果，顯示有多少預訂參加了該課程。使用視覺化效果可以：

a.說明參與的變數的哪個類別具有最多的觀測值

b.解釋觀測值是否在所參加變數的類別之間平衡

```python
# 这段代码的作用是进行数据可视化，它使用了 matplotlib 和 seaborn 库来创建一个柱状图，展示了每个类别的出勤次数。

# 这里，.astype(str) 被用来确保 df['category'] 和 df['attended'] 列中的数据类型是字符串。这对于绘图是有用的，特别是如果这些列原本不是字符串类型的话。
    # 将'category'和'attended'列的数据类型转换为字符串
df['category'] = df['category'].astype(str)
df['attended'] = df['attended'].astype(str)

# 这部分代码计算了每个类别的出现次数，并将结果存储在一个新的DataFrame category_counts 中。它也将结果按照出现次数进行排序，以便在绘制柱状图时可以按照从最高到最低的顺序显示类别。   
    # 计算每个类别的数量，并重置索引以获得一个DataFrame
category_counts = df['category'].value_counts().reset_index()
    # 重命名列以便于理解
category_counts.columns = ['category', 'count']
    # 按照'count'列的值降序排列
category_counts = category_counts.sort_values(by='count', ascending=False)

# 这里创建了一个柱状图，显示了每个类别的出勤次数。hue='attended' 参数添加了一个分类的维度，区分了出席和未出席。order 参数指定了柱状图的顺序，这里使用了之前按计数排序的类别
    # 设置图形的大小     
plt.figure(figsize=(15, 6))
    # 使用seaborn库的countplot函数来创建柱状图
ax = sns.countplot(data=df, x='category', palette=["#ebc1b7", "#9ac8ed"], hue='attended', order=category_counts['category'].astype(str))

# 这段代码遍历每个柱子（patches），并在每个柱子的上方添加一个文本标注来显示其高度，即该类别的计数。
    # 在每个柱状图上方添加文本标注显示其高度
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# 这里设置了坐标轴的标签和图形的标题，添加了图例，并且使用 sns.despine 移除了图形上方和右侧的轴边界，最后展示了这个图形。这些步骤都是为了改善图形的美观性和可读性。
    # 设置x轴和y轴的标签、图形的标题以及图例
plt.xlabel('Category')
plt.ylabel('')
plt.title('Count of Attendance by Category', size=14, weight="bold")
plt.legend(title='Attended', labels=['Not Attended', 'Attended'])
    # 移除上方和右侧的轴边界
sns.despine(left=True, bottom=True)
    # 显示图形
plt.show()
```

```python
# 这段代码利用了Pandas和Seaborn的功能，通过一个直观的水平堆叠条形图展示了不同类别的出勤百分比。这是数据分析和机器学习中数据可视化的常见做法，帮助理解数据的分布和模式。

# 这里，pd.crosstab() 函数创建了一个交叉表，它显示了df['category']（类别）与df['attended']（出席情况）之间的关系。参数normalize='index'表示按行归一化，即每行的和为1（或100%），这样你就可以得到每个类别的出席和未出席的百分比。结果乘以100将这些比例转换为百分比格式，并且用round()函数四舍五入到一位小数。
    # 创建一个类别和出席情况的交叉表，并将百分比四舍五入到一位小数
ct = round(pd.crosstab(df['category'], df['attended'], normalize='index') * 100,1)

# plt.figure() 设置了图形的尺寸。ct.plot() 是Pandas DataFrame的方法，用于直接从DataFrame生成图形。参数kind='barh'表示绘制水平条形图，stacked=True表示条形图是堆叠的，color参数定义了条形的颜色。
    # 设置图形的大小
plt.figure(figsize=(15, 5))
    # 使用DataFrame的plot方法绘制堆叠条形图 
ax = ct.plot(kind='barh', stacked=True, color= ["#ebc1b7", "#9ac8ed"])

# 这些代码设置了y轴标签为'Category'，x轴标签为空，以及标题为'Percentage of Attendance by Category'。plt.legend() 设置了图例，title参数指定了图例标题，labels参数指定了图例中各项的标签，bbox_to_anchor参数设定了图例的位置。
    # 设置y轴标签、x轴标签为空和图形标题
plt.ylabel('Category')
plt.xlabel("")
plt.title('Percentage of Attendance by Category', size=14, weight="bold")
    # 设置图例，并将其放置在图形的右上角
plt.legend(title='Attended', labels=['Not Attended', 'Attended'], bbox_to_anchor=(1, 1))

# ax.containers 包含了条形图中的所有条形（每个类别的出席和未出席）。ax.bar_label() 方法为每个条形添加标签，label_type='center' 表示标签位于条形的中心，fontsize和color参数控制标签的字体大小和颜色。
    # 为每个条形添加文本标签，显示数值
for container in ax.containers:
    ax.bar_label(container, label_type='center', fontsize=9, color='black')

# 最后，sns.despine() 函数用于移除图形周围的边框，plt.show() 显示最终的图形。
    # 移除图形上方和右侧的轴线
sns.despine(left=True, bottom=True)
    # 显示图形
plt.show()
```

```python
# 这段代码用于识别和定义异常值（离群点）的边界，这是机器学习数据预处理中一个常见的步骤。异常值可能会对某些机器学习模型的性能产生不利影响，特别是那些对异常值敏感的模型，如线性回归。
# 在机器学习中，处理异常值是重要的数据清洗步骤，因为异常值可能扭曲模型训练过程，导致过度拟合或其他不良影响。通过识别和处理异常值，我们可以提高模型的鲁棒性和预测能力。在实际操作中，处理异常值可能包括移除它们、进行变换以降低它们的影响，或者在某些情况下保留它们，因为它们可能代表了重要的、合法的变化。处理异常值的策略取决于数据的性质和分析的目标。

# 这两行代码使用 .quantile() 方法来计算 'months_as_member' 列的第一四分位数（Q1）和第三四分位数（Q3）。quantile(0.25) 返回了数据中所有值中最小的25%的分割点，而 quantile(0.75) 返回了最小的75%的分割点。
    # 计算四分位数
Q1 = df['months_as_member'].quantile(0.25)
Q3 = df['months_as_member'].quantile(0.75)

# IQR（四分位间距）是第三四分位数和第一四分位数之差，它是衡量统计分布散度的一个指标。IQR用于描述中间50%数据的范围。
    # 计算四分位间距（IQR）
IQR = Q3 - Q1

# 这两行代码定义了异常值的下界和上界。任何小于 lower_bound 或大于 upper_bound 的值都被认为是异常值。这是基于一个常用的规则，称为“1.5xIQR规则”，它来源于箱线图中使用的相同逻辑。这个规则假设数据分布近似为正态分布，而数据集中的大部分值（大约99.3%）应该在这个范围内。超出这个范围的值通常被认为是潜在的异常值。
    # 定义异常值的上下界
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# 这段代码继续了上一段关于异常值处理的内容，接下来使用定义的异常值边界来过滤原始数据集中的异常值，并使用更新后的数据集创建一个没有异常值的箱线图。
# 这段代码的目的是在移除异常值后探索 'attended' 与 'months_as_member' 之间的关系。移除异常值有助于更准确地理解大多数数据的趋势，因为异常值可能会歪曲统计度量，如平均值和中位数。在机器学习中，正确处理异常值对于建立准确的预测模型非常重要。

# 在这里，df['months_as_member'] >= lower_bound 是一个条件，它检查 df 中 'months_as_member' 列的每个值是否大于或等于之前计算出的下界（lower_bound）。类似地，df['months_as_member'] <= upper_bound 检查是否小于或等于上界（upper_bound）。这两个条件都是布尔表达式，为每行生成 True 或 False。
# 如果 'months_as_member' 的值在 lower_bound 和 upper_bound 之间，这意味着它不是一个异常值，这两个条件都会是 True。
# 如果 'months_as_member' 的值小于 lower_bound 或大于 upper_bound，这意味着它是一个异常值，至少有一个条件会是 False。
# & 运算符是逐元素的布尔“与”运算，它只有在两个条件都为 True 时，才会返回 True。因此，只有当 'months_as_member' 的值既不低于下界也不高于上界时，相应的行才会被包含在新的数据框 df1 中。
# 简而言之，df1 包含了没有异常值的 'months_as_member' 数据，您可以在不受这些极端值影响的情况下继续分析或建模。
    # 根据计算出的异常值边界过滤数据
df1 = df[(df['months_as_member'] >= lower_bound) & (df['months_as_member'] <= upper_bound)]

# 这行代码使用 matplotlib 设置了接下来要生成的图形的大小，宽度为8英寸，高度为6英寸。
plt.figure(figsize=(8, 6))

# 此行代码使用 seaborn 的 boxplot 函数从过滤后的数据集 df1 创建一个箱线图。箱线图将 'attended' 列作为x轴（分类变量），'months_as_member' 列作为y轴（数值变量）。orient='v' 参数指定了箱线图是垂直方向的。palette 参数设置了箱线图的颜色。
sns.boxplot(data=df1, x='attended', y='months_as_member', orient='v',palette = ["#ebc1b7", "#9ac8ed"])

# 这些行代码设置了y轴标签（但实际上被重复设置并最终置为空字符串）和x轴标签。这里有一个小错误，因为 plt.ylabel 被调用了两次，最后一次没有传递任何参数，所以y轴标签将不会显示。
plt.ylabel('Number of Months as a Member')
plt.xlabel('Attendance')
plt.ylabel("")

# 这行代码设置了图形的标题，并调整了字体大小和粗细。size=12 指定了标题的字体大小，weight="bold" 表示标题文字为粗体。
plt.title('Relationship between Attendance and Number of Months (Without Outliers)', size=12, weight="bold")

# seaborn 的 despine 函数用于移除图形上方和右侧的轴线，使图形看起来更加简洁。然而，参数 left=True 和 bottom=True 通常不是必须的，因为默认情况下 despine 会移除顶部和右侧的边框。如果您想保留x轴和y轴的轴线，您应该移除这些参数或者设置为 top=True, right=True。
sns.despine(left=True, bottom=True)

# 显示图形。
plt.show()
```

```python
#這段程式碼是用來評估機器學習模型（在這個例子中是邏輯回歸分類器）在測試數據上的性能，並將結果以熱圖的形式視覺化。


#predict 函數是 scikit-learn 機器學習模型的一部分，用於進行預測。在這裡，regression_model 是一個已經用訓練數據訓練好的模型，它使用 predict 方法根據測試數據集 X_test 來預測結果。預測的結果被存儲在 y_predict 中。
y_predict = regression_model.predict(X_test)
#score 函數計算模型的準確率，這是模型預測正確的標籤數除以總標籤數的比例。這裡，X_test 是特徵數據集的測試集，y_test 是對應的真實標籤。score 方法返回一個浮點數，表示模型的準確率，並且被打印出來。
print(regression_model.score(X_test, y_test))
#confusion_matrix 函數是 scikit-learn 中的一個功能，用於計算混淆矩陣。混淆矩陣是一個方陣，其中每行代表實際的類別，每列代表預測的類別。labels 參數定義了矩陣中類別的順序。在這個案例中，我們有兩個類別 0 和 1。
cm0 = metrics.confusion_matrix(y_test, y_predict,labels=[0, 1])

#這行代碼使用 pandas 库將混淆矩陣 cm0 轉換成一個 DataFrame 對象，這樣可以更方便地對數據進行處理和視覺化。index 和 columns 參數分別為混淆矩陣提供了行名和列名
    # 将混淆矩阵转换为 DataFrame
df_cm0 = pd.DataFrame(cm0, index = ["No","Yes"], columns = ["No","Yes"])

#subplots 函數創建了一個圖形 (fig) 和一個子圖軸 (ax)。figsize 參數設定了圖形的大小。
#ax.matshow 是 matplotlib 中的一個函數，用於繪製矩陣或二維數組，常用於繪製混淆矩陣的熱圖。cmap 是色彩映射，plt.cm.Blues 是 matplotlib 預設的一種藍色調色彩映射，用於熱圖的顏色變化。
    # 創建一個熱圖
fig, ax = plt.subplots(figsize=(7,5))
cax = ax.matshow(df_cm0, cmap=plt.cm.Blues)

#這是一個 for 循環，使用 np.ndenumerate() 函數來遍歷 df_cm0.values，這是將 df_cm0 轉換為 NumPy 數組的值。
#(i, j) 是行索引和列索引。
#val 是在索引 (i, j) 處的單元格中的數值。
#ax.text() 函數用於在坐標 (j, i) 處的單元格中添加文本 val。
#va='center' 和 ha='center' 分別設置垂直對齊和水平對齊為居中，意味著文本將在單元格內居中顯示。
#通過這種方式，每個單元格中都會顯示其對應的數值，這在熱圖中非常有用，因為它可以讓觀察者快速準確地看到每個類別的預測結果數量。
    # 為每個單元格添加文本標註
for (i, j), val in np.ndenumerate(df_cm0.values):
    ax.text(j, i, f"{val}", va='center', ha='center')

#這裡的 ax.set_xticklabels() 和 ax.set_yticklabels() 函數用於設置熱圖的 x 軸和 y 軸上的刻度標籤。
#首先，[''] 是一個包含空字符串的列表，它在這裡用於將刻度的起始位置留空，因為 matplotlib 的刻度是從 0 開始的，但我們通常想從 1 開始顯示標籤。
#df_cm0.columns.tolist() 將 df_cm0 的列名轉換為列表。
#df_cm0.index.tolist() 將 df_cm0 的索引轉換為列表。
#這樣，坐標軸刻度標籤就會與混淆矩陣的列名和索引名相對應。
    # 添加坐標軸標籤
ax.set_xticklabels([''] + df_cm0.columns.tolist())
ax.set_yticklabels([''] + df_cm0.index.tolist())

# 顯示顏色條
plt.colorbar(cax)
plt.show()

# 打印混淆矩陣的數據框以查看其結構和數據。
print(df_cm0)  
```

```python
# 这段代码通过计算和输出三个关键的性能评估指标（精确度、召回率和F1分数）来评估分类模型的性能。这些指标提供了对模型性能的全面理解，尤其在处理不平衡数据集时非常重要。

# 精确度（Precision）precision = precision_score(y_test, y_predict)
    # 函数: precision_score 是 scikit-learn 库中用于计算精确度的函数。
    # 定义: 精确度是指在所有被模型预测为正类的实例中，实际为正类的比例。计算公式为 TP / (TP + FP)，其中 TP 是真阳性的数量，FP 是假阳性的数量。
    # 意义: 精确度衡量了模型在预测正类样本时的准确性。
precision = precision_score(y_test, y_predict)

# 召回率（Recall）recall = recall_score(y_test, y_predict)
    # 函数: recall_score 是 scikit-learn 库中用于计算召回率的函数。
    # 定义: 召回率（也称为敏感度或真阳性率）是指在所有实际为正类的实例中，被模型正确预测为正类的比例。计算公式为 TP / (TP + FN)，其中 FN 是假阴性的数量。
    # 意义: 召回率衡量了模型识别出所有正类样本的能力。
recall = recall_score(y_test, y_predict)

# F1分数（F1-Score）f1 = f1_score(y_test, y_predict)
    # 函数: f1_score 是 scikit-learn 库中用于计算 F1 分数的函数。
    # 定义: F1分数是精确度和召回率的调和平均数，计算公式为 2 * (precision * recall) / (precision + recall)。
    # 意义: F1分数是精确度和召回率的平衡指标，特别适用于不平衡数据集的情况。它在两者之间提供了一个平衡的度量，是评价模型综合性能的重要指标。
f1 = f1_score(y_test, y_predict)

# 最后，这些 print 函数调用将计算出的精确度、召回率和F1分数打印到控制台上。这可以让你直观地看到模型的性能指标。
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)

#通过这些指标，我们不仅能评估模型的整体准确性，还能深入了解模型在预测正类（如欺诈检测中的“欺诈”类）时的表现，尤其是在负类实例数量远多于正类实例的不平衡数据集中。这有助于识别模型在实际应用中可能面临的问题，如高召回率但低精确度，或者反之。
```

```python
# 这段代码通过计算和输出三个关键的性能评估指标（精确度、召回率和F1分数）来评估分类模型的性能。这些指标提供了对模型性能的全面理解，尤其在处理不平衡数据集时非常重要。

# 精确度（Precision）precision = precision_score(y_test, y_predict)
    # 函数: precision_score 是 scikit-learn 库中用于计算精确度的函数。
    # 定义: 精确度是指在所有被模型预测为正类的实例中，实际为正类的比例。计算公式为 TP / (TP + FP)，其中 TP 是真阳性的数量，FP 是假阳性的数量。
    # 意义: 精确度衡量了模型在预测正类样本时的准确性。
precision = precision_score(y_test, y_predict)

# 召回率（Recall）recall = recall_score(y_test, y_predict)
    # 函数: recall_score 是 scikit-learn 库中用于计算召回率的函数。
    # 定义: 召回率（也称为敏感度或真阳性率）是指在所有实际为正类的实例中，被模型正确预测为正类的比例。计算公式为 TP / (TP + FN)，其中 FN 是假阴性的数量。
    # 意义: 召回率衡量了模型识别出所有正类样本的能力。
recall = recall_score(y_test, y_predict)

# F1分数（F1-Score）f1 = f1_score(y_test, y_predict)
    # 函数: f1_score 是 scikit-learn 库中用于计算 F1 分数的函数。
    # 定义: F1分数是精确度和召回率的调和平均数，计算公式为 2 * (precision * recall) / (precision + recall)。
    # 意义: F1分数是精确度和召回率的平衡指标，特别适用于不平衡数据集的情况。它在两者之间提供了一个平衡的度量，是评价模型综合性能的重要指标。
f1 = f1_score(y_test, y_predict)

# 最后，这些 print 函数调用将计算出的精确度、召回率和F1分数打印到控制台上。这可以让你直观地看到模型的性能指标。
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```
